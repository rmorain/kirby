```python
import pandas as pd
import sqlite3
import time

conn = sqlite3.connect("Database/wiki_database2.db")


'''
Tables organization: 
Entities_[letter or not_alpha]
entity_id: Primary key, Unique 
label: name of the entity
description: Short description 

Properties_relations Table
entity_id: relationship of  
property_id: Key for properties table
related_entity_id

Aliases Table
entity_id 
alias: Other labels. 

Properties Table
property_id Unique primary key  
label:Description of the Property 


'''

def exit_precedure():
    conn.close()
    exit(-1)


def get_table_name(entity_label):
    if entity_label[0].isalpha():
        return entity_label[0]
    else:
        return 'not_alpha'


def get_similar_entities_id(entity_label):
    entities_id = None
    table_name = get_table_name(entity_label)
    try:
        entities_id = pd.read_sql_query(
            "SELECT entity_id FROM Entities_{} WHERE label  LIKE \"%{}%\";".format(table_name, entity_label),
            conn)
    except Exception as e:
        print(e)
        exit_precedure()
    return entities_id


def get_specific_entity_id(entity_label):
    entity_id = None
    table_name = get_table_name(entity_label)
    try:
        entity_id = pd.read_sql_query(
            "SELECT entity_id FROM Entities_{} WHERE label = \"{}\";".format(table_name, entity_label),
            conn)
    except Exception as e:
        print(e)
        exit_precedure()
    return entity_id


def get_entity_properties(entity_label):
    entity_properties = None
    try:
        entity_properties = pd.read_sql_query(
            "SELECT * FROM Properties_relations WHERE entity_id = \"{}\";".format(entity_label),
            conn)
    except Exception as e:
        print(e)
        exit_precedure()
    return entity_properties


def get_entity_label(entity_id):
    relation_label = None
    try:
        relation_label = pd.read_sql_query(
            "SELECT label FROM Entities WHERE entity_id = \"{}\"".format(entity_id),
            conn)
    except Exception as e:
        print(e)
        exit_precedure()
    if relation_label.empty:
        return None
    return relation_label.iloc[0]["label"]


def get_property_label(property_id):
    entity_label = None
    try:
        entity_label = pd.read_sql_query(
            "SELECT label FROM Properties WHERE property_id = \"{}\"".format(property_id),
            conn)
    except Exception as e:
        print(e)
        exit_precedure()
    if entity_label.empty:
        return None
    return entity_label.iloc[0]["label"]


def get_relation_string(entity_id, property_id, related_entity_id):
    # assert(isinstance(relation, str) and len(relation) == 3)
    # relation_entity = relation
    entity_label = None
    related_entity_label = None
    property_name = None
    try:
        entity_label = get_entity_label(entity_id)
        related_entity_label = get_entity_label(related_entity_id)
        property_name = get_property_label(property_id)
    except Exception as e:
        print(e)
        exit_precedure()
    if entity_label is None or  related_entity_label is None or property_name is None:
        return None
    return entity_label + property_name + related_entity_label


# user_input = ''

# while len(user_input) == 0:
#      user_input= input("Please enter a entity's id: ")
#      property_id = input("Please enter property id: ")
#      related_entity_id = input("Please enter related entity's id: ")


# start_time = time.time()
# print(get_relation_string(user_input, property_id, related_entity_id))
# print("Relationship string took {} seconds".format(time.time() - start_time))
# # exit()
# # e_id = get_specific_entity_id(user_input).iloc[0]["entity_id"]
# # print("Select query on table Entities_{} took: {}".format(user_input[0], time.time() - start_time))
# # print(e_id)
# # # print(type(e_id))
# # # print(e_id.iloc[0]["entity_id"])
# # start_time = time.time()
# # e_properties = get_entity_properties(e_id)
# # print("Select query on table Properties_relations took: {}".format(time.time() - start_time))
# # print(e_properties)

# conn.close()



```
